Let's import and preview the data we will use to illustrate how to visualize data.

```python
import pandas as pd
vehicles = pd.read_csv("vehicles.csv")
vehicles.head()
```

## How to create a Relationship Visualization

Usually a scatterplot chart. And it shows how one variable is changing in response to another.
On the Pandas dataFrame we can call the "plot" method and as arguments call in the specifications for the chart.

```python
vehicles.plot(kind="scatter", x="citympg", y="co2emissions")
```

## How to create a Distribution Visualization

Shows the features' statistic distribution. Like histogram. But since this is useful for only one column, we need to select that column fromt he DF

```python
vehicles["co2emissions"].plot(kind="hist")
```

## How to create a Comparision Visualization

The most commonly used type is the Boxplot.

## How to create a Composition Visualization

This shows the ingridients within the DF and how it is added to the whole.
For this a few steps are required.
- Frist create an aggregarion.

```python
vehicles.groupby("year")["drive"].value_counts()
```

```python
vehicles.groupby('year')['drive'].value_counts().unstack()
```

```python
vehicles.groupby('year')['drive'].value_counts().unstack().plot(kind="bar", stacked= True,figsize=(10,6))
```
